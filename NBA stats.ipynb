$$ \ We \ access \ active \ players \ basic \ informations, \ each \ one \ of \ them \ has \ a \ unique \ id, \ no \ matches \ will \ return \ a \ null \ value.\\
\ We \ notice \ that \ there \ are \ 519 \ active \ players. $$

```python
from nba_api.stats.static import players
# get_players returns a list of dictionaries, each representing a player.
nba_players = players.get_active_players()
print('Number of players fetched: {}'.format(len(nba_players)))
nba_players[:5]
```

$$ \ As \ we \ can \ see \ below, \ we \ can \ draw \ a \ player \ with \ his \ id \ as \ a \ return \ we \ get \ his \ statistics \ from \ various \ categories \\
\ for \ all \ the \ years \ he \ is \ playing \ in \ the \ National \ Basketball \ Association. $$

```python
from nba_api.stats.endpoints import playercareerstats
stats = playercareerstats.PlayerCareerStats(player_id=1629638)
stats.career_totals_regular_season.get_data_frame()
```

$$ \ Below \ we \ select \ all \ 519 \ active \ players \ using \ sleep \ fuction \ otherwise \ we \ have \ a \ denied \ access.\\
\ Each \ player \ is \ exported \ in \ a \ dataframe \ of \ his \ own \ and \ the \ total \ procedure \ takes \ approximately \ 9 \ minutes. $$

```python
## import time
import pandas as pd
import time

dataframes = []
for player in nba_players[:519]:
    stats = playercareerstats.PlayerCareerStats(player_id=player["id"])
    dataframes.append(stats.career_totals_regular_season.get_data_frame())
    print('Got dataframe for player ' + str(player["id"]))
    time.sleep(1)

dataframes
```

$$ \ Below, \ one \ can \ notice \ that \ we \ connect \ all \ those \ dataframes \ and \ we \ successfully \\ \ have \ the \  \ total \ advanced \ statistics \ of \ all \ active \ NBA \ players. $$

```python
data500 = pd.concat(dataframes)
data500.to_csv('NBAstats.csv', index=False)
data500
```
